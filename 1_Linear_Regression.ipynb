```
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
diabetes = load_diabetes()
x = diabetes.data[:, 2]
y = diabetes.target
```

```
class Neuron:
    def __init__(self):
        self.w = 1.
        self.b = 1.

    def forward(self, x):
        y_hat = self.w * x + self.b
        return y_hat
    
    def backprop(self, x, err):
        w_grad = x * err
        b_grad = 1 * err
        return w_grad, b_grad
    
    def fit(self, x, y, epochs=1):
        for i in range(epochs):
            for x_i, y_i in zip(x, y):
                y_hat = self.forward(x_i)
                
                err = -(y_i - y_hat)
                w_grad, b_grad = self.backprop(x_i, err)
                
                self.w -= w_grad
                self.b -= b_grad
        
    def draw(self, x, y):
        plt.scatter(x,y)
        pt1 = (-0.1, neuron.forward(-0.1))
        pt2 = (0.15, neuron.forward(0.15))
        plt.xlabel('x')
        plt.ylabel('y')
        plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
        plt.show()
```

```
neuron = Neuron()
neuron.draw(x, y)
```

```
neuron.fit(x, y)
neuron.draw(x, y)
```
